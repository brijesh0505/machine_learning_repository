```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.show()
```

```python
%matplotlib inline
```

```python
sns.countplot(x='Survived',data=df)
```

```python
survived_counts = df[df['Survived'] == 1]['Sex'].value_counts()

# Create a bar plot to visualize the number of survivors for each gender
plt.figure(figsize=(8, 6))
sns.barplot(x=survived_counts.index, y=survived_counts.values)
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Count of Survived and Not Survived Passengers by Gender')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title('Count of Survived and Not Survived Passengers by Gender')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
titanic_age_dropped = df['Age'].dropna()

# Create a displot of the age distribution
plt.figure(figsize=(8, 6))
sns.displot(titanic_age_dropped, kde=True, bins=30)
plt.title('Age Distribution in Titanic Dataset')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='SibSp',data=df)
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Pclass',y='Age',data=df,palette='winter')
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return 37

        elif Pclass == 2:
            return 29

        else:
            return 24

    else:
        return Age
```

```python
df['Age'] = df[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```
