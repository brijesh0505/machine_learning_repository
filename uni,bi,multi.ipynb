```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

# Single figure with different colors for each species
sns.scatterplot(data=df, x="petal length (cm)", y="petal width (cm)", hue="species", palette="Set1", alpha=0.7)
plt.title('Sepal Length vs Sepal Width with Different Species')
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

# Pair plot for multivariate analysis
sns.set(style="ticks")
sns.pairplot(df, hue="species", palette="Set1")
plt.show()
```
